# SQL hodgepodge

## Table numrows and size MB

```sql
SELECT
s.name AS SchemaName,
t.name AS TableName,
MAX(p.rows) AS RowCounts,
(SUM(a.total_pages) * 7) / 1024.0 as TotalSpaceMB,
(SUM(a.used_pages) * 7) / 1024.0 as UsedSpaceMB,
(SUM(a.data_pages) * 7) /1024.0 as DataSpaceMB
FROM sys.tables t
JOIN sys.schemas s ON t.schema_id = s.schema_id
INNER JOIN sys.indexes i ON t.OBJECT_ID = i.object_id
INNER JOIN sys.partitions p ON i.object_id = p.OBJECT_ID AND i.index_id = p.index_id
INNER JOIN sys.allocation_units a ON p.partition_id = a.container_id
WHERE i.OBJECT_ID > 254
AND i.index_id IN (-1,1)
-- AND t.name LIKE 'TranslationTable%'
GROUP BY s.name, t.name
ORDER BY TotalSpaceMB DESC
```

## XML PATH

```sql
SELECT STUFF(
    (
    SELECT CONCAT(' UNION SELECT ', xi.COLUMN_NAME, ' FROM ', xi.TABLE_NAME)
    FROM INFORMATION_SCHEMA.COLUMNS xi
    WHERE xi.COLUMN_NAME LIKE '%ADDR_CDE%'
    FOR XML PATH ('')
    ),
    1, 7, ''
) x;
```

## Generic pivot

```sql
-- The key-value table.
CREATE TABLE #keyvalue (k INT, v VARCHAR(16), o INT);
INSERT INTO #keyvalue VALUES
(193, 'POLBA', 1),
(201, 'PHED', 1),
(208, 'EDSMA', 2),
(208, 'JOJO', 1),
(208, 'ENGBA', 3),
(208, 'HUMMA', 4);

-- Variables.
DECLARE @script NVARCHAR(MAX);
DECLARE @keyname VARCHAR(256) = 'id';
DECLARE @valuename VARCHAR(256) = 'major';
DECLARE @orderby VARCHAR(256) = 'o'

-- Create a rownum for the table.
DECLARE @rownumscript VARCHAR(128) = CONCAT('SELECT k AS ', @keyname, ', v, CONCAT(''', @valuename, '-'', ROW_NUMBER() OVER (PARTITION BY k ORDER BY ', @orderby, ')) valuenum FROM #keyvalue');

-- Get distinct values num.
DECLARE @distinctvalues NVARCHAR(MAX);
SET @script = CONCAT('WITH base AS ( ', @rownumscript, ' ),
xmlpath(distinctvalues) AS ( SELECT DISTINCT CONCAT(''['', valuenum, ''],'') FROM base FOR XML PATH ('''') )
SELECT @distinctvalues = SUBSTRING(distinctvalues, 1, LEN(distinctvalues)-1) FROM xmlpath');
EXEC sp_executesql @script, N'@distinctvalues NVARCHAR(MAX) OUTPUT', @distinctvalues = @distinctvalues OUTPUT;

SET @script = CONCAT('
	SELECT * FROM (', @rownumscript, ') x
	PIVOT
	(
		MAX(v)
		FOR valuenum IN (', @distinctvalues, ')
	) p;
');
EXEC sp_executesql @script;
DROP TABLE #keyvalue;
```

## Add rownum

## Rows to CSV

```sql
DECLARE @sql NVARCHAR(MAX) = NULL;

-- Populate the temp table.
;WITH base AS (
    SELECT xo.ID,
        CONCAT(
            '( ',
            STUFF((
            SELECT CONCAT(', ''', y.element, '''')
            FROM TWU_ANOM_EXCEPTION xi
            CROSS APPLY dbo.TWU_FN_STRING_SPLIT(xi.KEY_VALUE, '|') y
            WHERE xo.ID = xi.ID
            FOR XML PATH ('')
            ), 1, 2, ''),
            ' )')
        AS insertrow
    FROM TWU_ANOM_EXCEPTION xo
    WHERE xo.CATEGORY = @CATEGORY
    GROUP BY xo.ID
)
SELECT @sql = N'INSERT INTO ' + @TEMP_TABLENAME + ' VALUES ' + STUFF((SELECT ', ' + insertrow FROM base FOR XML PATH ('')), 1, 2, '');
```

## Search syscomments

```sql
SELECT CONCAT(schema_name(o.schema_id), '.', object_name(c.id)) objname, text
FROM sys.syscomments c
JOIN sys.all_objects o ON c.id = o.object_id
WHERE text LIKE '%BestPersonEmail%'
ORDER BY objname;
```
